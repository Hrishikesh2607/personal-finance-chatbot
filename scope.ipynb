## User Persona
My chatbot is built for students/young professionals who want to track monthly spending and get instant budget advice using natural language.

## Sample Conversation
“How much did I spend on food this month?”

“Set my food budget to 6000”

“Can I afford a movie this weekend?”

“How much money do I have left?”

“Show my monthly summary”

“I spent 250 on travel today”

“Am I overspending on food?”

“What’s my biggest expense?”

“I earn ₹40,000 per month — how much should I spend on food?”

“What is the difference between a credit card and a debit card?”


## One-Sentence Problem Statement
Build an NLP-based personal finance chatbot that helps users understand spending limits, budgeting, and basic financial concepts to improve money management habits.

## Feature List
DO:

Answer basic finance questions(budgeting, savings, ddebit, credit terms).

Help users estimate spending limits by category(food, rent, travel, etc).

Provide simple budgeting advice based on user inputs.

Explain financial terms in plain language.

Suggest healthy financial habits (saving tips, expense control).

DON'T:

Do not give professional investment or legal advice.

Do not access real bank accounts or sensitive personal data.

Do not make promises about profits or financial outcomes.

Do not store personal financial data without consent.
